```python
import numpy as np
import matplotlib.pyplot as plt

import pickle
import torch

%matplotlib inline
```

#### Load the data from pickle

```python
with open("train_sample.pkl", "rb") as f:
    data = pickle.load(f)
```

#### Convert data to image [0, 255]

```python
def to_data(x):
    """ Converts variable to numpy.
    """
    if torch.cuda.is_available():
        x = x.cpu()
    x = x.data.numpy()
    x = ((x + 1) * 255 / (2)).astype(np.uint8)
    return x
```

#### Draw the Samples

```python
def view_samples(samples):
    iteration = len(samples)
    rows = iteration // 5
    for i in range(0, iteration, 5):
        fig, axes = plt.subplots(1, 5, figsize=(12,6))
        for j in range(5):
            img = np.transpose(to_data(samples[i+j].squeeze(0)), (1,2,0))
            axes[j].imshow(img)
            axes[j].set_axis_off()
    plt.show();
                
```

```python
view_samples(data)
```
